```python
Median_Age = pd.read_csv("data/median-age.csv")
```

```python
Median_Age.info()
```

```python
Median_Age_continent_Africa = Median_Age[Median_Age["Entity"] == "Africa"]
Median_Age_continent_Africa.drop(columns = "Code",inplace = True)
Median_Age_continent_Africa
```

```python
Median_Age_continent_Europe = Median_Age[Median_Age["Entity"] == "Europe"]
Median_Age_continent_Europe.drop(columns = "Code",inplace = True)
Median_Age_continent_Europe
```

```python
Median_Age_continent_Asia = Median_Age[Median_Age["Entity"] == "Asia"]
Median_Age_continent_Asia.drop(columns = "Code",inplace = True)
Median_Age_continent_Asia
```

```python
Median_Age_continent_Latin = Median_Age[Median_Age["Entity"] == "Latin America & Caribbean"]
Median_Age_continent_Latin.drop(columns = "Code",inplace = True)
Median_Age_continent_Latin
```

```python
Median_Age_continent = pd.concat([Median_Age_continent_Africa, Median_Age_continent_Europe, Median_Age_continent_Oceania, Median_Age_continent_US,
          Median_Age_continent_Asia, Median_Age_continent_Latin], ignore_index=True)

Median_Age_continent
```
